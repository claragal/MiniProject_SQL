```python
import pandas as pd

df = pd.read_csv('/Users/danielebelmiro/Downloads/Conjunto.csv', sep=';')

print(df.head())
```

```python
column_mapping = {
    "idno": "respondent_id",
    "cntry": "country",
    "gndr": "gender",
    "yrbrn": "birth_year",
    "hhmmb": "household_members",
    "chldhhe": "children_in_house",
    "marital": "marital_status",
    "maritala": "marital_status_a",
    "hinctnt": "household_income",
    "hinctnta": "household_income_a",
    "pdwrk": "paid_work",
    "rtrd": "retired",
    "uempla": "unemployed_active",
    "uempli": "unemployed_inactive",
    "cmsrv": "community_service",
    "hswrk": "housework",
    "dsbld": "disabled",
    "edctn": "education_status",
    "dngoth": "activity_other",
    "dngref": "activity_refusal",
    "dngdk": "activity_unknown",
    "dngna": "activity_no_answer",
    "netuse": "internet_use",
    "netusoft": "internet_freq",
    "netustm": "internet_time",
    "trstsci": "trust_scientists",
    "impdiff": "value_new_experiences",
    "impdiffa": "value_new_experiences_a",
    "iphlppl": "value_help_others",
    "iphlppla": "value_help_others_a",
    "happy": "happiness",
    "health": "health_status",
    "rlgdgr": "religiosity",
    "sclact": "social_activities",
    "sclmeet": "social_meetings",
    "inmdisc": "discuss_matters",
    "name": "dataset_title",
    "inwtm": "interview_length",
    "inwyr": "interview_year",
    "mode": "collection_mode",
    "blgetmg": "minority_ethnic_group",
    "dscrgrp": "discriminated_group",
    "edulvla": "highest_education_level"
}

df.rename(columns=column_mapping, inplace=True)

df.head()
```

```python
# 1. Tabela: respondents
df_respondents = df.filter(items=["respondent_id"])
df_respondents.to_csv("respondents.csv", index=False)

# 2. Tabela: sociodemographics
sociodemographics_columns = [
    "respondent_id", "country", "gender", "birth_year", "household_members", 
    "children_in_house", "marital_status", "marital_status_a", "household_income", 
    "household_income_a", "paid_work", "retired", "unemployed_active", 
    "unemployed_inactive", "community_service", "housework", "disabled", 
    "education_status", "activity_other", "activity_refusal", 
    "activity_unknown", "activity_no_answer", "highest_education_level"
]
df_sociodemographics = df.filter(items=sociodemographics_columns)
df_sociodemographics.to_csv("sociodemographics.csv", index=False)

# 3. Tabela: social_identity
social_identity_columns = ["respondent_id", "minority_ethnic_group", "discriminated_group"]
df_social_identity = df.filter(items=social_identity_columns)
df_social_identity.to_csv("social_identity.csv", index=False)

# 4. Tabela: internet_use
internet_use_columns = ["respondent_id", "internet_use", "internet_freq", "internet_time"]
df_internet_use = df.filter(items=internet_use_columns)
df_internet_use.to_csv("internet_use.csv", index=False)

# 5. Tabela: opinions_values
opinions_values_columns = [
    "respondent_id", "trust_scientists", "value_new_experiences", 
    "value_new_experiences_a", "value_help_others", "value_help_others_a"
]
df_opinions_values = df.filter(items=opinions_values_columns)
df_opinions_values.to_csv("opinions_values.csv", index=False)

# 6. Tabela: personal_wellbeing
personal_wellbeing_columns = ["respondent_id", "happiness", "health_status"]
df_personal_wellbeing = df.filter(items=personal_wellbeing_columns)
df_personal_wellbeing.to_csv("personal_wellbeing.csv", index=False)

# 7. Tabela: religion_social_life
religion_social_life_columns = [
    "respondent_id", "religiosity", "social_activities", 
    "social_meetings", "discuss_matters"
]
df_religion_social_life = df.filter(items=religion_social_life_columns)
df_religion_social_life.to_csv("religion_social_life.csv", index=False)

# 8. Tabela: survey_metadata
survey_metadata_columns = [
    "metadata_id", "respondent_id", "dataset_title", 
    "interview_length", "interview_year"
]
df_survey_metadata = df.filter(items=survey_metadata_columns)
df_survey_metadata.to_csv("survey_metadata.csv", index=False)
```
